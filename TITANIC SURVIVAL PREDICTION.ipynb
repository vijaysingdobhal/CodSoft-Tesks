Handling Missing Values

```python
#Fill missing values in Age with the median age 
data['Age'].fillna(data['Age'].median(), inplace = True)
```

```python
#Convert cabin to a binary feature indicating whether a cabin is know
data['Cabin'] = data['Cabin'].notnull().astype(int)
```

```python
#Fill missing values in Embarked with the most common value
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace = True)
```

Drop Irrelevant Columns

```python
data.drop(['PassengerId','Name','Ticket'], axis = 1, inplace = True)
```

Convert Categorical Variables to Numerical

```python
#Convert Sex to a numerical veriable
data['Sex'] = data['Sex'].map({'male':0, 'female': 1})
```

```python
data
```

```python
# one- hot encode 'Embarked' and the drop first category to avoid multicollinearity
data = pd.get_dummies(data,columns = ['Embarked'], drop_first = True)
```

```python
data.head()
```

Feature Engineering

```python
# Create a new feature 'FamilySize'
data['FamilySize'] = data['SibSp'] + data['Parch']
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Saparate features and target variable
x = data.drop('Survived', axis = 1)
y = data['Survived']
```

```python
# Split the data into training (80%) and testing (20%) sets
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

Choose a Model and Train it

```python
from sklearn.linear_model import LogisticRegression
```

```python
#inialize the model
model = LogisticRegression(max_iter=200)
```

```python
# Train the model
model.fit(x_train, y_train)
```

Make Predictions,
use trained model to make predictions on the test set

```python
#Predict on the test set
y_pred = model.predict(x_test)
```

Evaluate the Model
1. Evaluate the model performace using metrics like accuracy, precision, recall and the F1 score.

```python
from sklearn.metrics import accuracy_score, precision_score,recall_score,f1_score,confusion_matrix
```

# Calcualte accuracy

```python
accuracy = accuracy_score(y_test,y_pred)
precision = precision_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)
f1 = f1_score(y_test,y_pred)
```

```python
# Print Results
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 score:", f1)
```

```python
# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)
```
